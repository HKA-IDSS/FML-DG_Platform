```python
import pandas as pd

adult_dataset = pd.read_csv("../../data/datasets/Adult/train.csv")
```

```python
adult_dataset["workclass"] = [string[1:] for string in adult_dataset["workclass"]]
adult_dataset["education"] = [string[1:] for string in adult_dataset["education"]]
adult_dataset["marital-status"] = [string[1:] for string in adult_dataset["marital-status"]]
adult_dataset["occupation"] = [string[1:] for string in adult_dataset["occupation"]]
adult_dataset["relationship"] = [string[1:] for string in adult_dataset["relationship"]]
adult_dataset["race"] = [string[1:] for string in adult_dataset["race"]]
adult_dataset["sex"] = [string[1:] for string in adult_dataset["sex"]]
adult_dataset["native-country"] = [string[1:] for string in adult_dataset["native-country"]]
adult_dataset["label"] = [string[1:] for string in adult_dataset["label"]]
```

```python
adult_dataset.loc[adult_dataset["age"] < 18, "age"] = 18
```

```python
adult_dataset = adult_dataset.replace("?", pd.NA)
```

```python
from sklearn.impute import SimpleImputer

adult_dataset = pd.DataFrame(SimpleImputer(missing_values=pd.NA, strategy='most_frequent').fit_transform(adult_dataset), index=adult_dataset.index, columns=adult_dataset.columns)
```

```python
adult_dataset.to_csv("Adult_ready_for_preprocessing.csv")
```
